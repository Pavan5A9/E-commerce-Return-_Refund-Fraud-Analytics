```python
# ------------------- IMPORT LIBRARIES -------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score

import joblib

# ------------------- LOAD DATASET -------------------
df = pd.read_csv("ecommerce_return_refund_fraud_dataset_100k.csv")
print("✅ Dataset Loaded")
print(df.head(), "\n")
print(df.info(), "\n")
print("Missing values:\n", df.isnull().sum())

# ------------------- EXPLORATORY DATA ANALYSIS -------------------
plt.figure(figsize=(15, 10))

# 1. Fraud Distribution
plt.subplot(2, 3, 1)
df["fraud_label"].value_counts().plot(kind="bar", color='skyblue')
plt.title("Fraud vs Legitimate")
plt.xlabel("Fraud Label")
plt.ylabel("Count")

# 2. Fraud Rate by Product Category
plt.subplot(2, 3, 2)
df.groupby("product_category")["fraud_label"].mean().plot(kind="bar", color='orange')
plt.title("Fraud Rate by Category")
plt.xticks(rotation=45)

# 3. Avg Return Ratio vs Fraud
plt.subplot(2, 3, 3)
df.groupby("fraud_label")["return_ratio"].mean().plot(kind="bar", color='green')
plt.title("Avg Return Ratio")

# 4. Fraud by Payment Method
plt.subplot(2, 3, 4)
df[df["fraud_label"]==1]["payment_method"].value_counts().plot(kind="bar", color='red')
plt.title("Fraud by Payment Method")
plt.xticks(rotation=45)

# 5. Delivery Status vs Fraud
plt.subplot(2, 3, 5)
pd.crosstab(df["delivery_status"], df["fraud_label"]).plot(kind="bar", ax=plt.gca())
plt.title("Fraud vs Delivery Status")
plt.xticks(rotation=0)

# 6. Account Age vs Fraud
plt.subplot(2, 3, 6)
df.groupby("fraud_label")["customer_account_age_days"].mean().plot(kind="bar", color='purple')
plt.title("Avg Account Age")

plt.tight_layout()
plt.show()

# ------------------- DATA PREPROCESSING -------------------
df_ml = df.drop(['order_id', 'customer_id'], axis=1)  # Remove ID columns

# Encode categorical columns
encoders = {}
for col in df_ml.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df_ml[col] = le.fit_transform(df_ml[col])
    encoders[col] = le

# Features & Target
X = df_ml.drop('fraud_label', axis=1)
y = df_ml['fraud_label']

# ------------------- FEATURE SELECTION -------------------
selector = RandomForestClassifier(n_estimators=50, random_state=42)
selector.fit(X, y)
top_6_features = pd.Series(selector.feature_importances_, index=X.columns).sort_values(ascending=False).head(6).index.tolist()
print("✅ Top 6 Features Selected:", top_6_features)

X_top = X[top_6_features]

# ------------------- MODEL TRAINING -------------------
X_train, X_test, y_train, y_test = train_test_split(X_top, y, test_size=0.2, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate model
y_pred = model.predict(X_test)
print(f"✅ Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
print("Classification Report:\n", classification_report(y_test, y_pred))

# ------------------- PREDICTION FUNCTION -------------------
def predict_fraud_status(user_input):
    missing = [f for f in top_6_features if f not in user_input]
    if missing:
        return f"❌ Error: Missing fields in input: {missing}", 0
    
    input_df = pd.DataFrame([user_input])
    
    # Encode categorical inputs
    for col in top_6_features:
        if col in encoders:
            label = input_df[col].iloc[0]
            if label in encoders[col].classes_:
                input_df[col] = encoders[col].transform([label])[0]
            else:
                input_df[col] = 0  # default for unknown labels

    prediction = model.predict(input_df[top_6_features])[0]
    probabilities = model.predict_proba(input_df[top_6_features])[0]
    
    status = "⚠️ FRAUD DETECTED" if prediction == 1 else "✅ NOT FRAUD"
    confidence = probabilities[prediction] * 100
    
    return status, confidence

# ------------------- EXAMPLE PREDICTION -------------------
my_input = {
    'alert_generated': 'Yes',
    'return_ratio': 0.75,
    'alert_severity': 'Medium',
    'total_returns': 50,
    'total_orders': 70,
    'refund_amount': 1500.0
}

result, conf = predict_fraud_status(my_input)
print(f"\nPrediction: {result}")
if conf > 0:
    print(f"Confidence: {conf:.2f}%")
```
